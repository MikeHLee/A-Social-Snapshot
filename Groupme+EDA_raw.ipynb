### Import Groupme JSON into Normalized Pandas DF

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
import json
from pandas.io.json import json_normalize

%matplotlib notebook
%matplotlib inline
plt.style.use
plt.rcParams['figure.figsize'] = (15, 10)
```

```python
# normalize the json into a df. 
with open('message.json') as data:
    dump = json.load(data)
    df = json_normalize(dump)
    
# send df to csv for Mike. 
df.to_csv('groupme_raw.csv')
```

```python
df
```

```python
# == CLEAN DATAFRAME ==  
#######################

# Cut out system generated messages like "so and so joined the channel"
df2 = df[df.user_id != 'system']

# Cut event data columns, almost all of which were NaNs. 

df2 = df2.drop('event.data.adder_user.id', axis=1)
df2 = df2.drop('event.data.adder_user.nickname', axis=1)
df2 = df2.drop('event.data.avatar_url', axis=1)
df2 = df2.drop('event.data.conversation.id', axis=1)
df2 = df2.drop('event.data.event.id', axis=1)
df2 = df2.drop('event.data.event.name', axis=1)
df2 = df2.drop('event.data.user.id', axis=1)
df2 = df2.drop('event.data.user.nickname', axis=1)
df2 = df2.drop('event.type', axis=1)
df2 = df2.drop('event.data.url', axis=1)
df2 = df2.drop('event.data.added_users', axis=1)
df2 = df2.drop('event.data.options', axis=1)
df2 = df2.drop('event.data.poll.expiration', axis=1)
df2 = df2.drop('event.data.poll.id', axis=1)
df2 = df2.drop('event.data.poll.subject', axis=1)
df2 = df2.drop('user_id', axis=1)

# Cut other uninteresting columns. 
df2 = df2.drop('system', axis=1)
df2 = df2.drop('sender_type', axis=1)
df2 = df2.drop('group_id', axis=1) 
```

```python
df2.to_csv('groupme_cleaned.csv')
```
